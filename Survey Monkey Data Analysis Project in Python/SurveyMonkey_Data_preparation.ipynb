# drop columns 

```python
dataset = data.copy(deep=True)
```

```python
column_to_drop = ['Start Date', 'End Date', 'Email Address','First Name', 'Last Name', 'Custom Data 1']
column_to_drop
```

```python
dataset = dataset.drop(columns=column_to_drop,axis=1)
```

```python
dataset.shape
```

```python
id_vars=list(dataset.columns[:8])
value_vars = list(dataset.columns[8:])
```

```python
dataset_melted= dataset.melt(id_vars=id_vars,value_vars=value_vars,var_name='Questions-SubQuestions',value_name='Answers')
```

```python
dataset_melted
```

```python
(dataset_melted['Respondent ID']==2658722536).sum()
```

```python
dataset_melted.shape
```

```python
Questions_Data = pd.read_excel("Data - Survey Monkey Output.xlsx",sheet_name='Question')
```

```python
Questions_Data
```

```python
Drop_Question_Columns =['Raw Question','Raw Sub-Question','Sub-Question']
```

```python
Questions = Questions_Data.copy(deep=True)
```

```python
Questions = Questions.drop(columns=Drop_Question_Columns,axis=1)
```

```python
Questions
```

```python
Questions[Questions['Question']=='Question 29']
```

```python
dataset_merged = pd.merge(left =dataset_melted,right=Questions,how='left',right_on='Question +SubQuestion',left_on='Questions-SubQuestions')
```

```python
dataset_merged
```

```python
# dataset_merged[dataset_merged['Question']=='Question 29']
```

```python
# dataset_merged.groupby('Question')['Respondent ID'].nunique().reset_index()
```

```python
responsent=dataset_merged.dropna(subset='Answers')
```

```python
responsent
```

```python
len(dataset_merged),len(responsent)
```

```python
responsent=responsent.groupby('Question')['Respondent ID'].nunique().reset_index()
```

```python
responsent
```

```python
responsent.rename(columns={'Respondent ID':'No. of Respondents'},inplace=True)
```

```python
responsent_merged = pd.merge(left =dataset_merged,right=responsent,how='left',right_on='Question',left_on='Question')
```

```python
responsent_merged
```

```python
responsent_merged.drop('Questions-SubQuestions',axis=1,inplace=True)
```

```python
responsent_merged.shape
```

```python
same_answer=dataset_merged
same_answer=same_answer.groupby(['Question +SubQuestion','Answers'])['Respondent ID'].nunique().reset_index()
same_answer.rename(columns={'Respondent ID':'Same Answer'},inplace=True)
same_answer.head(20)
```

```python
Final_merged = pd.merge(left =responsent_merged,right=same_answer,how='left',right_on=['Question +SubQuestion','Answers'],left_on=['Question +SubQuestion','Answers'])
Final_merged['Same Answer'].fillna(0,inplace=True)
```

```python
Final_merged
```

```python
Final_merged.rename(columns={'Identify which division you work in.-Response':'Primary Division',
       'Identify which division you work in.-Other (please specify)':'Secondary Division',
       'Which of the following best describes your position level?-Response':'Position',
       'Which generation are you apart of?-Response':'Generation',
       'Please select the gender in which you identify.-Response':'Gender',
       'Which duration range best aligns with your tenure at your company?-Response':'Tenure-Duration',
       'Which of the following best describes your employment type?-Response':'Employment-Type'},inplace=True)
```
