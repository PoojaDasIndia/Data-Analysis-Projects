## <span style="color:#FC8019;font-size:25px; font-style:bold;"><u><b>About Dataset</b></u></span>

Context:

This dataset contains swiggy registered restaurants details of major metropoliton cities of India. It considered only metropoliton cities with population 4.5 million. As per the Census of India 2011 definition of more than 4 million population, some of the major Metropolitan Cities in India are:


Content:

This Dataset Contains 18 Different Columns which are :

 1.  type --> Contain (F,D,C,P) Means 
         > Fine Restaurants
         > Dessert Restaurants
         > Continental Restaurants
         > Premium Restaurants
 
 2.   id --> Restaurants id                  
 3.   name  --> Restaurants name           
 4.   uuid --> Universally Unique Identifiers
 5.   city -->  Restaurants Belong from City  
 
            * Mumbai (more than 18 Million)
            * Delhi (more than 16 Million)
            * Kolkata (more than 14 Million)
            * Chennai (more than 8.6 million)
            * Bangalore (around 8.5 million)
            * Hyderabad (around 7.6 million)
            * Ahmedabad (around 6.3 million)
            * Pune (around 5.05 million)
            * Surat (around 4.5 million)
            
 6.   area  -->  Restaurants belong from Area              
 7.   avgRating --> Restaurants   Average Rating         
 8.   totalRatingsString --> Restaurants   Total  Rating  string
 9.   cuisines  --> Restaurants   Serving Cuisines (Menu)
 10.  costForTwoStrings  --> Restaurants   Cost for Two 
 11.  deliveryTime    --> Restaurants   Delivery Time
 12.  minDeliveryTime  --> Restaurants  Minimum Delivery Time
 14.  maxDeliveryTime  --> Restaurants  Maximum Delivery Time  
 14.  address  --> Restaurants  address 
 15.  locality  --> Restaurants  locality
 16.  unserviceable --> Restaurants  provide service or not
 17.  veg    --> Restaurants   Serving food veg. or non-veg
 18.  City

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Copy original Dataset</b></u></span> 

```python
dataset_copy=dataset.copy(deep=True)
```

```python
dataset_copy.head()
```

## <span style="color:#FC8019;font-size:25px; font-style:bold;"><u><b>Data Cleaning Process start from here</b></u></span> 

```python
dataset_copy['type']
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Rename Column name 'type' to 'Restaurants Type'</b></u></span> 

```python
dataset_copy.rename(columns = {'type':'Restaurants Type'}, inplace = True)
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Define  Function for changing data name inside Restaurants type</b></u></span> 

```python
def replace(data):
    if data =='F':
        return data.replace("F","Fine")
    elif data =='D':
        return data.replace("D","Dining")
    elif data =='C':
        return data.replace("C","Continental")
    else:
        return data.replace("P","Premium")
```

```python
# apply function with the help of lambda
Type = dataset_copy['Restaurants Type'].apply(lambda x: replace(x))
```

```python
dataset_copy['Restaurants Type'] = Type
dataset_copy.head(5)
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Replace space ' ' between two cuisins with (,)</b></u></span> 

```python
cuisines = dataset_copy['cuisines'].apply(lambda x: x.replace(" ",","))
dataset_copy['cuisines'] = cuisines
dataset_copy.head(5)
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Rename Column name 'veg' to 'Vegetarian'</b></u></span> 

```python
dataset_copy.rename(columns = {'veg':'Vegetarian'}, inplace = True)
```

```python
dataset_copy.head(5)
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Drop Columns which not usefull like 'uuid' & 'City'</b></u></span> 

```python
dataset_copy.drop(columns=['uuid','City'], inplace=True)
```

```python
dataset_copy.head(2)
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Removing unnecessary string only keep the integer</b></u></span> 

```python
costForTwoStrings = dataset_copy['costForTwoStrings'].apply(lambda x: x.replace(x[-8:],'')[1:])
dataset_copy['costForTwoStrings'] = costForTwoStrings
dataset_copy.head(5)
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Removing Rating or rating  string  from the column</b></u></span> 

```python
totalRatingsString = dataset_copy['totalRatingsString'].apply(lambda s : s.replace(s[-8:],''))
dataset_copy['totalRatingsString'] = totalRatingsString
dataset_copy.head(2)
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Replacing '--' with '0.0' so that type change will be easy</b></u></span> 

```python
dataset_copy['avgRating'].isna()
```

```python
avgRating = dataset_copy['avgRating'].apply(lambda s : s.replace('--','0.0'))
dataset_copy['avgRating'] = avgRating
dataset_copy.head()
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Changing data type of  'avgRating' & 'costForTwoStrings' </b></u></span> 

```python
dataset_copy['avgRating'] = dataset_copy['avgRating'].astype('float')
```

```python
dataset_copy['costForTwoStrings'] = dataset_copy['costForTwoStrings'].astype('int64')
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Remove Outliers cost less than 2</b></u></span> 

```python
dataset_copy = dataset_copy[dataset_copy['costForTwoStrings']>2]
dataset_copy
```

```python
dataset_copy.info()
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Dropping Null values from 'unserviceable' & 'Vegetarian' </b></u></span> 

```python
dataset_copy.dropna(subset = ['unserviceable','Vegetarian','address','locality'],axis=0,inplace=True)
```

## <span style="color:#FC8019;font-size:20px; font-style:bold;"><u><b>Drop Duplicates if any present </b></u></span> 

```python
dataset_copy = dataset_copy.drop_duplicates()
```

```python
dataset_copy.isna().sum()
```

```python
dataset_copy.shape
```

```python
dataset_copy.to_csv('Swiggy_Final_Data.csv',index=False)
```
